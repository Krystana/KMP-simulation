```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
monday = pd.read_csv('data/monday.csv', delimiter =';')
tuesday = pd.read_csv('data/tuesday.csv', delimiter =';' )
wednesday = pd.read_csv('data/wednesday.csv', delimiter =';')
thursday = pd.read_csv('data/thursday.csv', delimiter =';')
friday = pd.read_csv('data/friday.csv', delimiter =';')
df = pd.concat([monday, tuesday, wednesday, thursday, friday], axis = 0)

df['timestamp'] = pd.to_datetime(df['timestamp'])
df['day'] = df['timestamp'].dt.day

#conditions

values = [4, 3, 5, 6]

#df['revenue_per_minute'] = np.select(conditions, values)

# df["unq_id"] = 'Cust' + df["day"].astype(str) + df["customer_no"].astype(str)

df["unq_id"] = 'Cust' + df["day"].astype(str) + df["customer_no"].astype(str)

df = df[['timestamp', 'day', 'unq_id', 'customer_no', 'location']] # 'day',

#df['time'] = df['timestamp'].dt.time

#df = df[['timestamp', 'unq_id', 'customer_no', 'location']] # 'time', 'day',

df.head()
```

```python
df = df.sort_values(['customer_no', 'timestamp'])
df = df.set_index('timestamp')
df.head()
# df.head(50)
# df.tail(20)
#df.tail(20)
```

```python
#reset index now to rename index to use it (because of duplication): 
df = df.reset_index(level = 1)
```

```python
df_temp = df
df_temp.index.rename('customers', inplace=True)
```

```python
df_temp
```

```python
# write first location for each customer to list
first_locations = df_temp.groupby('unq_id').location.first().to_list()
```

```python
# create dataframe out of list 
df_firsts = pd.DataFrame(first_locations)
```

```python
# get value counts in percentage == probabilty for next after entrance to dictionary
entrance_p ={}
for idx, loc in enumerate (df_firsts[0].value_counts(normalize=True).index.to_list()): 
    
    entrance_p[loc] = (df_firsts[0].value_counts(normalize=True)[idx]).round(3)
```

```python
entrance_p['location'] = 'entrance'
```

```python
entrance_p
```

```python
# add entrance_p to table: 
trans_prob_2 = trans_prob.reset_index().append(entrance_p, ignore_index=True)
```

```python
trans_prob_2 = trans_prob_2.set_index('location') #.reset_index('location')
```

```python
trans_prob_2.to_csv('trans_prob_2.csv', index=True)
```
