```python
def simulate(self):
        """Simulate the customer behavior n times
        Exit when customer ends up in checkout"""
    customer_trajectory = [self.state]
    while True:
        customer_trajectory.append(self.change_state())
        if self.state == 'checkout':
            break
        return customer_trajectory
```
