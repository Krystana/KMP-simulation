```python
import numpy as np



class Customer:
    """Customer class for Supermarket simulation"""
    
    
    def __init__(self, customer_name, transition_matrix): #state
        self.customer_name = customer_name
        self.state = 'entrance'
        self.transition_matrix = transition_matrix    # transition matrix
        
        
    def initial_state(self):

        #self.state = 'entrance'
        print(f' Customer starts at {self.state}')
        
        
    def __repr__(self):
        return f'Customer {self.customer_name} is in the {self.state} section'
        
    
    
    def change_state(self):
            
        self.state = np.random.choice(self.transition_matrix.columns.values, p=self.transition_matrix.loc[self.state]) #, replace=False)
        print:(f'Customer {self.customer_name} is in the {self.state} section')
```

## create class Supermarket

```python
import time
import datetime
from faker import Faker



fake = Faker()
open_supermarket = datetime.datetime(year= 2021, month=1, day=1,hour=7, minute=0)
close_supermarket = datetime.datetime(year= 2021, month=1, day=1,hour=21, minute=0)


class Supermarket:
    """manages multiple Customer instances that are currently in the market.
 """

    def __init__(self, customers): 
        self.customers = customers
        self.current_time = 0 
        
        
    def start_at_entrance(self): 
        """ prints the initial state for each customer w/ initial_state"""
    
        for element in customers:
            element.initial_state()
            print(f'{element}')                

    def get_time(self):  #open supermarket
        """ assing the current time to the opening hour of the supermarket """
  
        
        self.current_time = open_supermarket  # opening hour
        return self.current_time
    
    
    def next_minute(self):
        """propagates all customers to the next state w/ change_state"""
        
        self.current_time = self.current_time + datetime.timedelta(minutes=1)       
        for element in self.customers:
            element.change_state()
            print(f'{element}')
       # return self.current_time   
    
    
    
    def remove_ex_customers(self):
        """remove customers that entered state 'checkout' from list and print it"""
        
        for element in self.customers:
            if element.state == 'checkout':
                self.customers.remove(element)
                print(f'{element} and is removed')
    
    
    def generate_cust_append(self):
        
        no_of_appends = np.random.randint(0,2)
        appended_cust = []
        for i in range(no_of_appends):
    
    
            c = Customer(fake.name(), transition_matrix)
            appended_cust.append(c)
    
        return appended_cust    
 
    def simulate(self):
        """ supermarket simulation that takes in a list of customers"""
        
   
        self.get_time()
        self.start_at_entrance()   # only customers from list start at entrance ??
        self.next_minute()
        
        
        while  close_supermarket > self.current_time > open_supermarket: 
                
                print(self.current_time)
                #self.start_at_entrance()
                self.next_minute()
                self.remove_ex_customers()
                
                c = Customer(fake.name(), transition_matrix)

                appended_cust = self.generate_cust_append()
                
                for append in appended_cust:
                    self.customers.append(append)   
                    
                for element in self.customers:        # should print starting state for new cust
                    
                    #element.initial_state()
                    if element.state == 'entrance':
                    #print('\n\n\n')    #f'{element}'
                        print(f'{element} and happy to start shopping')
                          
                          
                #if self.current_time > close_supermarket:
                    
                    #break
   
```
