```python
import numpy as np

class Customer:
    """Customer class for Supermarket simulation"""
    
    def __init__(self, customer_no, state, transition_matrix):
        self.customer_no = customer_no
        self.state = state
        self.transition_matrix = transition_matrix    # transition matrix
        
    def __repr__(self):
        return f'Customer {self.customer_no} is in the {self.state} section'
         
        
    
    
    
    def change_state(self):
            
            #print(self.state)
            
               
            self.state = np.random.choice(self.transition_matrix.columns.values, p=self.transition_matrix.loc[self.state]) #, replace=False)
            
   # random.choices
```

## create class Supermarket

```python
class Supermarket:
    """manages multiple Customer instances that are currently in the market.
 """

    def __init__(self, customers):
        self.customers = customers
        #self.state = state
   #def get_time():
     #"""current time in HH:MM format,
     #"""
    
    def next_minute(self):
        """propagates all customers to the next state 
        w/ change_state
     """
        for element in self.customers:
            element.change_state()
            
    
    def remove_ex_customers(self):
        """remove churned from list"""
        
        for element in self.customers:
            if element.state == 'checkout':
                self.customers.remove(element)
                print(f'{element} is removed')
    
    
    #def add_new_customers(self, customer_no):
    #""" new list of customers ready to change state """
    def simulate(self):
        while len(self.customers) != 0:
            self.next_minute()
            self.remove_ex_customers()
            c = Customer(1, 'drinks', transition_matrix)
            self.customers.append(c)
            #print(len(self.customers))
            
        
        
    

   
```

```python
customers = [Customer(i, 'fruit',transition_matrix) for i in range(100)]
```

```python
s = Supermarket(customers)
```

```python
s.simulate()
```
